```python
#Get neuropixel experiment table
#This is where the data will be downloaded, if you don't have the data already. So put this manifest file where you want the data downloaded
manifest_path = './data/manifest.json'
cache = EcephysProjectCache.from_warehouse(manifest=manifest_path)

#These are the sessions that have enough runnning and rest data to perform the data analysis
sessions_of_interest = [778998620,789848216,839068429,839557629,840012044,771160300,778240327]
sessions = cache.get_session_table()
sessions.loc[sessions_of_interest]
```

```python
#Let's just stick with one for now
session_ID = 778998620

#These are the start times for the decoding window
start_times = np.concatenate((np.arange(-0.15,-0.1,0.01),np.arange(-0.10,0.10,0.005),np.arange(.10,.26,0.01)))
#100ms moving window 
tWindow = 0.1
stop_times = start_times + tWindow
nWindows = start_times.shape[0]; nKfold = 5; nClasses = 4

#Allocate arrays for timecourse of decoding
decoding_tc = np.zeros((nWindows,2,nKfold,nClasses,nClasses))
decoding_z_tc = np.zeros((nWindows,2,nKfold,nClasses,nClasses))
decoding_shf_tc = np.zeros((nWindows,2,nKfold,nClasses,nClasses))
CI95_shf_tc = np.zeros((nWindows,2,nKfold,2,nClasses,nClasses))
```

```python
print('\nSession {}: '.format(session_ID))
SaveDir = './results/decoding/s{}'.format(session_ID)

if not os.path.isdir(SaveDir):
    os.makedirs(SaveDir)

#Get session data; this is where the data will be downloaded if it isn't already
session = cache.get_session_data(session_ID)

#Let's collect all of the neurons across all areas
units_df = session.units

#Get running data during the drifting_gratings_75_repeats stimulus & drifting_gratings_contrast stimuli 
dgrepeats_df,_ =  util.determine_running_trials(session,tStart=-0.5, tEnd=.51,tOffset=0.5, stimulus='drifting_gratings_75_repeats')
dgcontrast_df,_ =  util.determine_running_trials(session,tStart=-0.5, tEnd=.51,tOffset=0.5, stimulus='drifting_gratings_contrast')

if session_ID == 778998620:
    stimulus_df = dgrepeats_df
else:
    #Combine each drifting grating stimulus table together
    stimulus_df = pd.concat([dgcontrast_df,dgrepeats_df])

#Combine trials into low vs high contrast types; 
#In the "75_repeats" block, drifting gratings are either presented at 0.1 or 0.8 contrast
#In the "contrast" block, drifting gratings are presented in 1 of 9 contrasts: [0.01, 0.02, 0.04, 0.08, 0.13, 0.2, 0.35, 0.6, 1.0]
#To keep the range of contrasts within each block small, we only used 5 of these
low_contrasts = [0.08,0.13,0.1,0.2]; high_contrasts = [0.6,0.8,1.0]

#For now let's just look at high contrast; uncomment / indent loop if desired
# for contrast_str in ['low','high']:
contrast_str = 'high'
target_contrasts = low_contrasts if contrast_str == 'low' else high_contrasts

#Match number of trials per condition and get indices
print('\t{}-contrast drifting gratings'.format(contrast_str))
trial_indices, trial_indices_nomatch, trial_indices_perstim = dc.get_trial_indices(stimulus_df,target_contrasts)

        
```

```python
uniq_areas = units_df['ecephys_structure_acronym'].unique()

vhs = [ 'VISp', 'VISl', 'VISal', 'VISpm', 'VISam']
areas_list = np.intersect1d(uniq_areas,vhs)

#For now let's just look at V1; uncomment / indent loop if desired
#     for area in areas_list:
area = 'VISp'
print('Constructing spike count matrix for {} ...'.format(area),end='\t')
tmpname = ccfsum.loc[ccfsum['abbreviation'] == area]['name'].values
areaname = tmpname[0] if len(tmpname) == 1 else area

#Get unit IDs in this area
area_uIDs = units_df.loc[units_df['ecephys_structure_acronym'] == area].index.values

## Format into spike counts per trial
tStart_spk = -0.50; tEnd_spk = 0.501; bSize_spk = 0.001; tOffset_spk = -0.5
time_bin_edges = np.arange(tStart_spk, tEnd_spk, bSize_spk)

#Format spike Counts into x-array
spike_counts_xr = session.presentationwise_spike_counts(
    bin_edges=time_bin_edges,
    stimulus_presentation_ids=stimulus_df.index.values,
    unit_ids = area_uIDs)

#Convert x-array to regular numpy
spike_counts = np.array(spike_counts_xr)
nTrials,nBins,nNeurons = spike_counts.shape
print('{} neurons'.format(nNeurons))
```

```python
#Do you want to match the population distributions between rest and running? 
match_distributions = False

#Pick your favorite classifier, or one of these ['LDA','SVM','Euclidean_Dist','nearest_neighbor']
classifier='Euclidean_Dist'

#For supplemental plots to understand the analysis
PlotDir = os.path.join(SaveDir,'plots')
if not os.path.isdir(PlotDir):
    os.makedirs(PlotDir)
    
#Decode within specific windows
decoding_tc[:] = 0
decoding_z_tc[:] = 0
decoding_shf_tc[:] = 0
CI95_shf_tc[:] = 0

#Run decoding analysis per window in parallel
timer_start = time.perf_counter()
parallel = True
if parallel:
    pool = mp.Pool(processes=15)
    processes = []
results = []

for tStart, tEnd in zip(start_times,stop_times):
    #Get indices for window
    iStart = int(np.round((tStart+np.abs(tOffset_spk))/bSize_spk))
    iEnd = int(np.round((tEnd+np.abs(tOffset_spk))/bSize_spk))

    #Slice spike count array
    spike_count_slice = spike_counts[:,slice(iStart,iEnd)]

    if match_distributions:
        fsuffix = 'mm_{}_{}-{}'.format(area,iStart,iEnd)
    else:
        fsuffix = '{}_{}-{}'.format(area,iStart,iEnd)
        
    #Perform decoding analysis
    if parallel:
        processes.append(pool.apply_async(dc.run_decoding_analysis,args=(spike_count_slice,stimulus_df,trial_indices,trial_indices_perstim,trial_indices_nomatch,match_distributions,PlotDir,fsuffix,classifier)))
    else:
        tmp = dc.run_decoding_analysis(spike_count_slice,stimulus_df,trial_indices,trial_indices_perstim,trial_indices_nomatch,match_distributions,PlotDir,fsuffix,classifier)
        results.append(tmp)

#Get results
if parallel:
    results = [p.get() for p in processes]
    pool.close()
print('done: {:.2f} minutes'.format((time.perf_counter()-timer_start)/60))

nSpikes_removed = []
for index,rTuple in enumerate(results):
    #Save results
    decoding_tc[index] = rTuple[0]
    decoding_shf_tc[index] = rTuple[1]
    decoding_z_tc[index] = rTuple[2]
    CI95_shf_tc[index] = rTuple[3]
    nSpikes_removed.append(rTuple[4])
```

```python
#Take mean across kfolds for plotting
dc_hits_tc = np.zeros((nWindows,2,nClasses))
shf_hits_tc = np.zeros((nWindows,2,nClasses))
shf_CI95_tc = np.zeros((nWindows,2,2))

for index,t in enumerate(start_times):
    for iR,runstr in enumerate(['rest','run']):
        dc_hits_tc[index,iR] = np.diag(np.mean(decoding_tc[index,iR],axis=0))
        shf_hits_tc[index,iR] = np.diag(np.mean(decoding_shf_tc[index,iR],axis=0))

        #Calculate 95% CI over kfolds
        shf_CI95_tc[index,iR,0] = np.mean(np.diag(np.mean(CI95_shf_tc[index,iR,:,0],axis=0)))
        shf_CI95_tc[index,iR,1] = np.mean(np.diag(np.mean(CI95_shf_tc[index,iR,:,1],axis=0)))

shf_hits_tc_mean =  np.mean(shf_hits_tc,axis=-1)
dc_hits_tc_mean = np.mean(dc_hits_tc,axis=-1)  
```

```python
#Calculate decoding speed and latencies between rest vs run
decoding_measures, delta_decoding_measures, fig = dc.analyze_decoding_timecourse(dc_hits_tc,shf_hits_tc,shf_CI95_tc,start_times,area,contrast_str)
```
